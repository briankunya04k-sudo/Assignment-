```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data={

      'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
      'Age': [25, 30, 35, 28],
      'Salary': [50000, 60000, 70000, 55000],
      'Department': ['HR', 'IT', 'Finance', 'HR']
}
```

```python
Brian=pd.DataFrame(data)
print('Original DataFrame')
print(Brian)
```

```python
print('\nFirst2 rows:')
print(Brian.head(2))
```

```python
print('\nBasic statistics:')
print(Brian.describe())
```

```python
name_salary=Brian[['Name','Salary']]
print('\nName and Salary columns:')
print(name_salary)
```

```python
#filter people older than 28
older_than_28=Brian[Brian['Age']>28]
print('\nPeople older than 28:')
print(older_than_28)
#
```

```python
#add new column'Bonus as 10% of salary
Brian['Bonus']=Brian['Salary']*0.1
print('\nDataFrame with Bonus column:')
print(Brian)
```
